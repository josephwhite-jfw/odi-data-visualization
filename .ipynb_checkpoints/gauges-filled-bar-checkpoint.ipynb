```python
import pathlib, re, numpy as np, pandas as pd, plotly.graph_objects as go
```

```python
# --- 1.  Explicit repo root ----------------------------------------------
ROOT_DIR = pathlib.Path(r"C:\Users\Owner\OneDrive\Documents\GitHub\odi-data-visualization")

DATA_DIR  = ROOT_DIR / "data"
STYLE_NAME = "gauges_filled_bar"
PLOTS_DIR  = ROOT_DIR / "plots" / STYLE_NAME

TEST_MODE  = True
TEST_DEPTS = ["Finance and Management"]   # ignored if TEST_MODE=False

# make sure output dir exists
PLOTS_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# ── performance bands ─────────────────────────────────────────────────────
BANDS = [
    (1.00, 3.29, "#e53935"),   # red
    (3.30, 3.49, "#fb8c00"),   # orange
    (3.50, 3.79, "#fdd835"),   # yellow
    (3.80, 4.00, "#43a047"),   # green
]
band_colour = lambda v, d="#999999": next((c for lo,hi,c in BANDS if lo<=v<=hi), d)
```

```python
# ── helpers ───────────────────────────────────────────────────────────────
SAFE = re.compile(r'[^A-Za-z0-9 _\-()]+')
clean = lambda s, n=80: SAFE.sub('_', s)[:n].strip()

NUMRE = re.compile(r"[-+]?\d*\.?\d+")
to_num= lambda s: float(NUMRE.search(str(s)).group()) if NUMRE.search(str(s)) else np.nan

def wrap(txt, width=35):
    words, line, lines = txt.split(), "", []
    for w in words:
        if len(line)+len(w)+1 > width:
            lines.append(line.rstrip()); line=""
        line += w+" "
    lines.append(line.rstrip())
    return "<br>".join(lines)
```

```python
# ── load survey data ──────────────────────────────────────────────────────
def load_long():
    csvs = list(DATA_DIR.glob("*.csv"))
    if not csvs:
        raise FileNotFoundError(f"No CSVs in {DATA_DIR}")

    def tidy(p):
        df = pd.read_csv(p)
        if df.columns[0]!="Outcome":
            df = df.rename(columns={df.columns[0]:"Outcome"})
        return df[df["Outcome"].str.lower().str.strip()!="outcome"]

    wide = pd.concat([tidy(p) for p in csvs], ignore_index=True)
    wide["Outcome"] = wide["Outcome"].str.strip()
    wide.columns    = wide.columns.str.strip()

    long = (wide
            .melt(id_vars="Outcome", var_name="Department", value_name="raw")
            .assign(Score=lambda d:d["raw"].apply(to_num))
            .dropna(subset=["Score"])
            .loc[lambda d:d["Score"].between(1,4)])
    return long

long = load_long()

print("📋 Departments:", long["Department"].unique().tolist())
```

```python
# ── build gauge (grey background bands + coloured bar) ────────────────────
def build_gauge(value, title):
    grey_steps = [{"range":[lo,hi], "color":"#e0e0e0", "line":{"width":2,"color":"white"}}
                  for lo,hi,_ in BANDS]
    fig = go.Figure(go.Indicator(
        mode   ="gauge+number",
        value  = value,
        title  = {"text":wrap(title), "font":{"size":16,"color":"#1f3d5a"}},
        number = {"font":{"size":38,"color":"#1f3d5a"}},
        gauge  = {
            "axis":{"range":[1,4],
                    "tickmode":"array",
                    "tickvals":np.linspace(1,4,7),
                    "tickfont":{"size":12}},
            "steps": grey_steps,
            "bar":  {"color": band_colour(value), "thickness":0.45},
        }))
    fig.update_layout(height=450,width=450,
                      margin=dict(t=80,b=20,l=20,r=20))
    return fig
```

```python
# ── save one gauge ────────────────────────────────────────────────────────
def save(dept, outcome, score):
    d_dir = PLOTS_DIR / clean(dept)
    d_dir.mkdir(parents=True, exist_ok=True)
    fig   = build_gauge(score, outcome)
    base  = clean(outcome)
    fig.write_html(d_dir/f"{base}.html", include_plotlyjs="cdn")
    fig.write_image(d_dir/f"{base}.png", scale=2)
    print("   •", dept, "--", base+".png")
```

```python
# ── generate gauges ───────────────────────────────────────────────────────
targets = TEST_DEPTS if TEST_MODE else long["Department"].unique()
for dept in targets:
    rows = long[long["Department"].str.casefold()==dept.casefold()]
    if rows.empty:
        print(f"⚠️  No data for {dept}"); continue
    print("Generating",dept)
    for _, r in rows.iterrows():
        save(dept, r["Outcome"], r["Score"])

print("\n✅ Files saved to", PLOTS_DIR)
```
